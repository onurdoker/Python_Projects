## House Price Statistical Analyzer

Subject: Statistical calculations and NumPy/Pandas filtering

Components: Pandas, NumPy, House Price Data (csv)

Requirements:
- Calculate mean, median and standard deviation values for all numerical columns
- Filter records belonging to a specific region code using NumPy masks or Pandas loc function
- Find the average house price on the selected region
- Create a Matplotlib Scatter Plot to show the relationship between two variables

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Loading house price dataset
df = pd.read_csv('housing.csv')
```

```python
# Calculate mean, median and standard derivation value
mean_values = df.mean(numeric_only=True)
median_values = df.median(numeric_only=True)
std_values = df.std(numeric_only=True)

print('Mean Values:')
print(mean_values)
print('\nMedian Values:')
print(median_values)
print('\nStandard Deviation Values:')
print(std_values)
```

```python
# Filter records belonging to a specific region code using NumPy masks or Pandas loc function
region_code = 4000
filtered_df = df.loc[df['area'] == region_code]

print(filtered_df.head())
```

```python
# Find the average house price on the selected region
print(f'The average house price in region {region_code} is {filtered_df['price'].mean():.2f}$')
```

```python
# Create a Matplotlib Scatter Plot to show the relationship between two variables
plt.figure(figsize=(10, 6))
plt.scatter(df['area'], df['price'])
plt.title(f'House Price vs Size')
plt.xlabel('House Size (sqft)')
plt.ylabel('Price ($)')
plt.grid(True)
plt.show()
```
