## Title: Titanic Data Analysis Report

Subject: Data manipulation with Pandas and Visualization with Seaborn

Components: Pandas, NumPy, Seaborn, Matplotlib, titanic.csv

Requirements:
- Load data, fill missing Age values with median
- Calculate survival rates by Gender and Pclass using groupby
- Create a Seaborn Countplot to show survival by Gender
- Create a Seaborn Boxplot to display the relationship between Pclass and Fare

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Loading Titanic Dataset
data = pd.read_csv('titanic.csv')
```

```python
# Display basic information about the dataset
print('Initial DataInformation:')
print(data.info())
```

```python
# Fill missing Age values with median age
median_age = data['age'].median()
data['age'] = data['age'].fillna(median_age)
```

```python
# Calculate survival rate by Gender and Pclass using groupby
survival_rates = data.groupby(['sex', 'pclass'])['survived'].mean().reset_index()
survival_rates.columns = ['Gender', 'Pclass', 'Survival Rate']
print('\nSurvival Rates by Gender and Pclass:')
print(survival_rates)
```

```python
# Create a Seaborn Countplot to show survival by Gender
plt.figure(figsize=(10, 10))
sns.countplot(data=data, x='sex', hue='survived')
plt.title('Survival Count by Gender')
plt.xlabel('Gender')
plt.ylabel('Count')
plt.legend(title='Survived', labels=['Not Survived', 'Survived'])
plt.show()
```

```python
# Create a Seaborn Boxplot to display the relationship between Pclass and Fare
plt.figure(figsize=(10, 6))
sns.boxplot(data=data, x='pclass', y='fare')
plt.title('Relationship between Pclass and Fare')
plt.xlabel('Pclass')
plt.ylabel('Fare')
plt.show()
```
